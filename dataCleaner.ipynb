```python
import os
import csv
```

```python
list_files = os.listdir("./assets")
num_files = len(list_files)

for file in list_files:
    file_path = "./assets/" + file
    total_str = ""
    print(file_path)
    if (os.path.isfile(file_path)):
        with open(file_path,encoding='utf-8') as input_file, open('cleanAssets/' + file , 'w', newline='') as output_file:
            reader = csv.reader(input_file)
            writer = csv.writer(output_file)
            for row in reader:
                if any(row):
                    row = row[:6]
                    row = [element.replace("-", "0") for element in row]
                    writer.writerow(row)
```
